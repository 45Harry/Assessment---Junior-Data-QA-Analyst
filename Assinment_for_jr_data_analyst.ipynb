# TASK  1
#### Please provide a manifest file for post_code, each unique DOB in that post_code and the 
#### total number of unique DOBs in that zip code, and the count of records in that 
#### post_code. Example: 
#### post_code: 5000 dob: 1959,1960, 2000, 2019 
#### total_unique_dob: 4 
#### total_records: 40 


```python
filtered_df.shape
```

```python
manifest_df = filtered_df.groupby('post_code') .agg(
    dob = ('dob', lambda x : sorted(x.unique().tolist())) , # Gives sorted unique dobs
    totol_unique_dob = ('dob','nunique'), # counts total unique dob
    total_records =('dob','count') # count total records
    ).reset_index()
manifest_df
```

```python
# Formated the output as requested
def format_row(row):
    dobs = ', '.join(map(str, row['dob']))
    return f"post_code: {row['post_code']}\ndob: {dobs}\ntotal_unique_dob: {row['totol_unique_dob']}\ntotal_records: {row['total_records']}\n"

# Creating the manifest content
manifest_content = '\n'.join(manifest_df.apply(format_row, axis=1))

```

```python
print(manifest_content)
```

```python
# Writing the manifest content into manifest file
with open('manifest.txt','w') as f:
    f.write(manifest_content)
```

# Task 2

```python
unique_post_code = filtered_df['post_code'].unique()

# Create a dictionary to store grouped DataFrames
grouped = dict(tuple(manifest_df.groupby('post_code')))

for post_code , group_df in grouped.items():
    filename = f'dataset_{(post_code)}.csv'  
    group_df.to_csv(f'Datasets/{filename}', index=False)
    print(f'Saved at: Datasets/{filename}')


print(f"\nSuccessfully created {len(unique_post_code)} files.")
```
